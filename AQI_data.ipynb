```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("/aqi data.xlsx")
df.tail(10)
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df["LOCATIONS"],df["AQI-US"])
plt.show()
```

```python
plt.figure(figsize=(15,6))
import plotly.express as px
fig=px.scatter_3d(df,x="AQI-US",y="STATUS",z="LOCATIONS", color=df["LOCATIONS"])
fig.show()
```

```python
plt.figure(figsize=(15,6))
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
import plotly.express as px
fig=px.scatter_3d(df,x="AQI-US",y="TEMP",z="HUMID", color=df["LOCATIONS"])
fig.show()
```

```python
plt.figure(figsize=(15,6))
import plotly.express as px
fig=px.scatter_3d(df,x="AQI-US",y="PM2.5",z="PM10", color=df["LOCATIONS"])
fig.show()
```
